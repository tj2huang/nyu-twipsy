```r
file <- 'E:/sept_fpl_ts.csv'
fpl <- read.csv(file)
```

```r
casual = fpl[, 'casual']
```

```r
casual = log(casual)
```

```r
casual_ts <- ts(casual, start=1, end=30, frequency = 24)
```

```r
plot(casual_ts, type='l')
```

```r
acf(casual_ts, lag.max=200)
```

```r
pacf(casual_ts, lag.max=200)
```

```r
plot(diff(casual_ts, 168))
```
